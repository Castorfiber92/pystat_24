### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

### Answer
* There are two approaches: either using pyplot ("MATLAB style") or OO (object oriented)
* The recomended way is OO.

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

### Answer
* The figure is the "whole". Everything is placed on the figure.
* Axes are regions for plotting data. When we start to draw plots, we will first create a figure and decide on the number of axes ('how many graphs do I want?')
* Axis sets the scale and limits for the plots.
* Artists: everything visible is an artist (all visible elements)

### When plotting in Matplotlib, what is the expected input data type?

### Answer

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
#OOP approach

x = np.linspace(-4, 4, 100)

fig, ax = plt.subplots()
ax.plot(x, x**2)
ax.set_xlabel('x - label')
ax.set_ylabel('y - label')
ax.set_title('OO title')
```

```python
x = np.linspace(-4, 4, 100)
#Pyploy approach
plt.plot(x, x**2)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Pyplot title')
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

fig, axs = plt.subplots(1, 2, figsize = (10, 4))
axs[0].scatter(random_data_x, random_data_y, color = 'Green')
axs[0].set_title('Scatterplot')

axs[1].set_xticks([0, 1, 2, 3, 4])
axs[1].set_xticklabels(names, rotation=45, ha='right')
axs[1].bar(names, values, color = ('Purple', 'Red', 'Orange', 'Yellow', 'Green'))
axs[1].set_title('Barplot')


```
